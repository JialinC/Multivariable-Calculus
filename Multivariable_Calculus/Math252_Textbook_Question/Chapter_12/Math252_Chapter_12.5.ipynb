Example.1

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sympy import *
t,s,x,y,z = symbols("t s x y z")
point = np.array([5,1,3])
vector = np.array([1,4,-2])
line = point + t*vector
print(line)
pointA = [i.subs(t,1) for i in line]
pointB = [i.subs(t,-1) for i in line]
print(pointA)
print(pointB)
```

Example.2

```python
A = np.array([2,4,-3]) 
B = np.array([3,-1,1])
AB = B - A
line = A + t*AB
print(line)
ans = solve(line[2],t)
intersection = [i.subs(t,ans[0]) for i in line]
print(intersection)
```

Example.4

```python
x,y,z = symbols("x y z")
P = (2,4,-1)
V = np.array([x,y,z]) - P
n = np.array([2,3,4])
plane = V.dot(n)
print(plane)
fig4 = plt.figure(figsize=(8,8))
ax4 = fig4.gca(projection='3d')
X = np.arange(0, 6, 0.25)
Y = np.arange(0, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = -0.5*X - 0.75*Y  + 3
ax4.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax4.set_xlabel('X axis')
ax4.set_ylabel('Y axis')
ax4.set_zlabel('Z axis')
ax4.view_init(azim=360)
ax4.set_xlim(0,6)
ax4.set_ylim(0,4)
ax4.set_zlim(0,3)
plt.show()
```

Example.5

```python
x,y,z = symbols("x y z")
P = np.array([1,3,2]) 
Q = np.array([3,-1,6])
R = np.array([5,2,0])
a = Q - P
b = R - P
n = np.cross(a,b)
V = np.array([x,y,z]) - P
plane = V.dot(n)
print(plane)
fig5 = plt.figure(figsize=(8,8))
ax5 = fig5.gca(projection='3d')
X = np.arange(1, 5, 0.25)
Y = np.arange(-1, 3, 0.25)
X, Y = np.meshgrid(X, Y)
Z = -6/7*X - 10/7*Y  + 50/7
ax5.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax5.set_xlabel('X axis')
ax5.set_ylabel('Y axis')
ax5.set_zlabel('Z axis')
ax5.view_init(azim=420)
ax5.set_xlim(0,5)
ax5.set_ylim(0,3)
ax5.set_zlim(0,6)
plt.show()
```

```python
t,x,y,z = symbols("t x y z")
n1 = np.array([1,1,1])
n2 = np.array([1,-2,3])
cos_theta = n1.dot(n2)/(np.linalg.norm(n1)*np.linalg.norm(n2))
theta = np.arccos(cos_theta)/np.pi*180
print(theta)
v = np.cross(n1,n2)
point = np.array([1,0,0])
line = point + t*v
print(line)
```

Example.8

```python
a,b,c,x0,y0,z0,x1,y1,z1 = symbols("a b c x0 y0 z0 x1 y1 z1")
P1 = np.array([x1,y1,z1])
P0 = np.array([x0,y0,z0])
n = np.array([a,b,c])
B = P1 - P0
D = abs(n.dot(B))/sqrt(n.dot(n))
print(D)
```
