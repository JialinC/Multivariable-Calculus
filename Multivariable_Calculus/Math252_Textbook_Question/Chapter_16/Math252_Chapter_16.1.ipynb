Example.1

```python
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(np.arange(-3, 3.5, .5), np.arange(-3, 3.5, .5))
U = -Y
V = X

fig = plt.figure(figsize=(8,8))
plt.quiver(X, Y, U, V, units='width')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

Example.2

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.array([-0.5, -0.1, 0.1,0.5]))

u = 0
v = 0
w = z

ax.quiver(x, y, z, u, v, w,length=0.4)

plt.show()
```

Example.4

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')
np.seterr(divide='ignore', invalid='ignore')
x, y, z = np.meshgrid(np.array([-1.0, -0.5, 0.0, 0.5, 1.0]),
                      np.array([-1.0, -0.5, 0.0, 0.5, 1.0]),
                      np.array([-1.0, -0.5, 0.0, 0.5, 1.0]))

u = -x/(x**2+y**2+z**2)**(3/2)
v = -y/(x**2+y**2+z**2)**(3/2)
w = -z/(x**2+y**2+z**2)**(3/2)

ax.quiver(x, y, z, u, v, w,length=0.2,normalize=True)

plt.show()
```

Example.6

```python
X, Y = np.meshgrid(np.arange(-4, 4.5, .5), np.arange(-4, 4.5, .5))
U = 2*X*Y
V = X**2-3*Y**2

fig = plt.figure(figsize=(8,8))
plt.quiver(X, Y, U, V, units='width')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```
