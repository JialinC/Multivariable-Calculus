```python
#a
import sympy as sp
x,y = sp.symbols("x y")
fa = sp.sqrt(x + y +1)/(x - 1)
fa32 = fa.subs([[x,3],[y,2]])
print(fa32)
```

```python
#b
fb = x*sp.log(y**2 - x)
fb32 = fb.subs([[x,3],[y,2]])
print(fb32)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig5 = plt.figure(figsize=(8,8))
ax5 = fig5.gca(projection='3d')
X = np.arange(0, 3, 0.05)
Y = np.arange(0, 4, 0.05)
X, Y = np.meshgrid(X, Y)
Z = -3*X - 2*Y  + 6
ax5.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax5.set_xlabel('X axis')
ax5.set_ylabel('Y axis')
ax5.set_zlabel('Z axis')
ax5.view_init(azim=30)
ax5.set_xlim(0,4)
ax5.set_ylim(0,5)
ax5.set_zlim(0,7)
plt.show()
```

Example.6

```python
fig6 = plt.figure(figsize=(8,8))
ax6 = fig6.gca(projection='3d')
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0,0.5*np.pi, 100)
radius = 3
X = radius * np.outer(np.cos(u), np.sin(v)) 
Y = radius * np.outer(np.sin(u), np.sin(v))
Z = radius * np.outer(np.ones(np.size(u)), np.cos(v))
ax6.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax6.set_xlabel('X axis')
ax6.set_ylabel('Y axis')
ax6.set_zlabel('Z axis')
ax6.view_init(azim=60)
ax6.set_xlim(-4,4)
ax6.set_ylim(-4,4)
ax6.set_zlim(0,4)
plt.show()
```

```python
x = np.arange(-5.0,5.0)
y = np.arange(-5.0,5.0)
x,y = np.meshgrid(x,y)
z = 6 - 3*x - 2*y
plt.figure(figsize=(8,8))
CS = plt.contour(x,y,z,[-6,0,6,12])
plt.clabel(CS)
plt.show()
```

```python
t = np.arange(0, 2*np.pi, 0.01)
X = np.cos(t)
Y = np.sin(t)
X,Y = np.meshgrid(X,Y)
Z = np.sqrt(9 - X**2 - Y**2)
plt.figure(figsize=(8,8))
CS = plt.contour(X,Y,Z)
plt.clabel(CS)
plt.show()
```
