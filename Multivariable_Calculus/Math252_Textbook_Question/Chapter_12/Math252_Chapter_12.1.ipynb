Example1.a

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig1_a = plt.figure(figsize=(8,8))
ax1_a = fig1_a.gca(projection='3d')
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 0*X + 0*Y + 3
ax1_a.set_xlabel('X axis')
ax1_a.set_ylabel('Y axis')
ax1_a.set_zlabel('Z axis')
ax1_a.set_zlim(0,6)
ax1_a.set_ylim(-4,4)
ax1_a.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```

```python
fig2_a = plt.figure(figsize=(8,8))
theta = np.arange(0, 2*np.pi, 0.01)
r = 1
X = r*np.sin(theta)
Y = r*np.cos(theta)
Z = 0*X + 0*Y + 3
ax2_a = fig2_a.gca(projection='3d')
ax2_a.plot(X, Y, Z)
ax2_a.set_xlabel('X axis')
ax2_a.set_ylabel('Y axis')
ax2_a.set_zlabel('Z axis')
ax2_a.set_xlim(-2,2)
ax2_a.set_ylim(-2,2)
ax2_a.set_zlim(0,6)
plt.show()
```

Example5

```python
#for example the center is (0,0,0) and r is 10
fig5 = plt.figure(figsize=(8,8))
ax5 = fig5.gca(projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
radius = 10
X = radius * np.outer(np.cos(u), np.sin(v)) 
Y = radius * np.outer(np.sin(u), np.sin(v))
Z = radius * np.outer(np.ones(np.size(u)), np.cos(v))
ax5.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax5.set_xlabel('X axis')
ax5.set_ylabel('Y axis')
ax5.set_zlabel('Z axis')
ax5.set_xlim(-15,15)
ax5.set_ylim(-15,15)
ax5.set_zlim(-15,15)
plt.show()
```

```python
fig6 = plt.figure(figsize=(8,8))
ax6 = fig6.gca(projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
radius = 2*np.sqrt(2)
X = radius * np.outer(np.cos(u), np.sin(v)) - 2 
Y = radius * np.outer(np.sin(u), np.sin(v)) + 3
Z = radius * np.outer(np.ones(np.size(u)), np.cos(v)) - 1
ax6.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax6.set_xlabel('X axis')
ax6.set_ylabel('Y axis')
ax6.set_zlabel('Z axis')
ax6.set_xlim(-5,1)
ax6.set_ylim(0,6)
ax6.set_zlim(-4,2)
plt.show()
```

```python
#for example the center is (0,0,0) and r is 10
fig7 = plt.figure(figsize=(8,8))
ax7 = fig7.gca(projection='3d')
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0.5*np.pi, np.pi, 100)
r = np.linspace(1, 2, 100) 
theta, r = np.meshgrid(u, r)
X1 = r*np.cos(theta)
Y1 = r*np.sin(theta)
Z1 = 0*theta
radius1 = 1
radius2 = 2
X = radius1 * np.outer(np.cos(u), np.sin(v)) 
Y = radius1 * np.outer(np.sin(u), np.sin(v))
Z = radius1 * np.outer(np.ones(np.size(u)), np.cos(v))
ax7.plot_surface(X, Y, Z, rstride=1, cstride=1,color='r')
X = radius2 * np.outer(np.cos(u), np.sin(v)) 
Y = radius2 * np.outer(np.sin(u), np.sin(v))
Z = radius2 * np.outer(np.ones(np.size(u)), np.cos(v))
ax7.plot_surface(X, Y, Z, rstride=1, cstride=1,color='r')
ax7.plot_surface(X1, Y1, Z1,color='g',alpha=0.7)
ax7.set_ylabel('Y axis')
ax7.set_zlabel('Z axis')
ax7.set_xlim(-3,3)
ax7.set_ylim(-3,3)
ax7.set_zlim(-3,3)
plt.show()
```
