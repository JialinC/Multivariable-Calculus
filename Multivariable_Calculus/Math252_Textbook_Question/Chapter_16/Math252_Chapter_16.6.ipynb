Example.1

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig1 = plt.figure(figsize=(8,8))
u = np.arange(0, 2*np.pi, 0.01)
r = 2
X = r*np.cos(u)
Y = np.array([-5, 5])
Z  = np.array([r*np.sin(u)])
X, Y = np.meshgrid(X, Y)
ax1 = fig1.gca(projection='3d')
ax1.plot_surface(X, Y, Z, rstride=1, cstride=1, color='b')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_zlabel('Z axis')
ax1.set_xlim(-3,3)
ax1.set_ylim(-5,5)
ax1.set_zlim(-3,3)
ax1.view_init(azim=30)
plt.show()
```

```python
fig4 = plt.figure(figsize=(8,8))
ax4 = fig4.gca(projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
radius = 1
X = radius * np.outer(np.cos(u), np.sin(v))
Y = radius * np.outer(np.sin(u), np.sin(v))
Z = radius * np.outer(np.ones(np.size(u)), np.cos(v))
ax4.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax4.set_xlabel('X axis')
ax4.set_ylabel('Y axis')
ax4.set_zlabel('Z axis')
ax4.set_xlim(-1.5,1.5)
ax4.set_ylim(-1.5,1.5)
ax4.set_zlim(-1.5,1.5)
plt.show()
```

```python
fig8 = plt.figure(figsize=(8,8))
ax8 = fig8.gca(projection='3d')
x = np.linspace(0, 2*np.pi, 200)
theta = np.linspace(0, 2*np.pi, 200)
X = np.outer(np.ones(np.size(x)),x)
Y = np.outer(np.sin(x),np.cos(theta))
Z = np.outer(np.sin(x),np.sin(theta))
ax8.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax8.set_xlabel('X axis')
ax8.set_ylabel('Y axis')
ax8.set_zlabel('Z axis')
ax8.set_xlim(-0,4)
ax8.set_ylim(-2,2)
ax8.set_zlim(-2,2)
ax8.view_init(azim=270)
plt.show()
```

```python
import sympy as sp
from sympy.physics.vector import *
R = ReferenceFrame('R')
theta,phi,a = sp.symbols("theta phi a")
r = a*sp.sin(phi)*sp.cos(theta)*R.x + a*sp.sin(phi)*sp.sin(theta)*R.y + a*sp.cos(phi)*R.z
rdphi = r.diff(phi,R)
rdtheta = r.diff(theta,R)
magnitude = sp.simplify(rdphi.cross(rdtheta).magnitude())
integrate = sp.integrate(sp.integrate(magnitude,(phi,0,sp.pi)),(theta,0,2*sp.pi))
print(integrate)
```

Example.11

```python
x,y,z,r,theta = sp.symbols("x y z r theta")
z = x**2 + y**2
zx = sp.diff(z,x)
zy = sp.diff(z,y)
f = sp.sqrt(1 + zx**2 + zy**2)
f = f.subs([[x,r*sp.cos(theta)],[y,r*sp.sin(theta)]])
ans = sp.integrate(sp.integrate(f*r,(r,0,3)),(theta,0,2*sp.pi))
print(ans)
```
