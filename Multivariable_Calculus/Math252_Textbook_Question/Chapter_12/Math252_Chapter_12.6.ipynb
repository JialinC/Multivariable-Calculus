Example.1

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig1 = plt.figure(figsize=(8,8))
ax1 = fig1.gca(projection='3d')
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 + 0*Y
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_zlabel('Z axis')
ax1.set_zlim(0,6)
ax1.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```

Example.3

```python
coefs = (1, 1/9, 1/4) 
rx, ry, rz = 1/np.sqrt(coefs)
fig3 = plt.figure(figsize=(8,8))
ax3 = fig3.gca(projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
radius = 2*np.sqrt(2)
X = rx * np.outer(np.cos(u), np.sin(v))
Y = ry * np.outer(np.sin(u), np.sin(v))
Z = rz * np.outer(np.ones(np.size(u)), np.cos(v)) 
ax3.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
ax3.set_xlabel('X axis')
ax3.set_ylabel('Y axis')
ax3.set_zlabel('Z axis')
ax3.set_xlim(-1,1)
ax3.set_ylim(-3,3)
ax3.set_zlim(-2,2)
ax3.view_init(azim=180)
plt.show()
```

```python
from __future__ import division
fig6 = plt.figure(figsize=(8,8))
ax6 = fig6.add_subplot(111, projection='3d')

r=1;
u=np.linspace(-1,1,200);
v=np.linspace(0,2*np.pi,60);
[u,v]=np.meshgrid(u,v);

a = 2
b = 1
c = 2

x = a*np.cosh(u)*np.cos(v)
y = b*np.cosh(u)*np.sin(v)
z = c*np.sinh(u)

ax6.plot_surface(x, y, z,  rstride=4, cstride=4, color='b')
ax6.set_ylabel('Y axis')
ax6.set_zlabel('Z axis')
ax6.set_xlim(-4,4)
ax6.set_ylim(-3,3)
ax6.set_zlim(-2,2)
ax6.view_init(azim=30)
plt.show()
```
