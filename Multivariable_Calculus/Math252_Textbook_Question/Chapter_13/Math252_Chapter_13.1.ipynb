```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig4 = plt.figure(figsize=(8,8))
t = np.arange(0, 20, 0.01)
X = np.sin(t)
Y = np.cos(t)
Z = t
ax = fig4.gca(projection='3d')
ax.plot(X,Y,Z)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.set_zlim(0,10)
plt.show()
```

Example.5

```python
from sympy.physics.vector import *
import sympy as sp
R = ReferenceFrame('R')
t = sp.symbols("t")
P = 1*R.x + 3*R.y - 2*R.z
Q = 2*R.x - 1*R.y + 3*R.z
r = (1 - t)*P + t*Q
print(r)
```

Example.6

```python
fig6 = plt.figure(figsize=(8,8))
t = np.arange(0, 2*np.pi, 0.01)
X = np.cos(t)
Y = np.sin(t)
Z = 2 - np.sin(t)
ax = fig6.gca(projection='3d')
ax.plot(X,Y,Z)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.set_zlim(0,4)
ax.view_init(azim=60)
plt.show()
```

```python
fig7 = plt.figure(figsize=(8,8))
t = np.arange(-2, 2, 0.01)
X = t
Y = t**2
Z = t**3
ax = fig7.gca(projection='3d')
ax.plot(X,Y,Z)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_xlim(-2,2)
ax.set_ylim(0,4)
ax.set_zlim(-6,6)
ax.view_init(azim=30)
plt.show()
```
